## Model Buliding

- Here splited data using Train Test Split by sklearn.model_selection importing train_test_split
- And importing Random Forest Classifier from sklearn.ensemble to Predict to score

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=10,criterion="entropy",random_state=0)
```

```python
x = df.drop('Attrition', axis=1)
y = df['Attrition']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
x_train.head() 
```

```python
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)
```

## Predicting for X_test

-  Predicting the Accuracy of the data by importing Accuracy score from sklearn.metrics

```python
pred=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pred)
```
